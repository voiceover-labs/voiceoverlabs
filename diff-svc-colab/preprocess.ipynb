## mp3 wav 파일로 변환

```python
import os
from pydub import AudioSegment

# 가수 이름 설정
singer_name = "sikyung"

# 변환할 파일이 있는 폴더 경로 설정 - 폴더 경로에 가수 이름 포함
folder_path = f""
output_path = f""

os.makedirs(output_path, exist_ok=True)

# 폴더 내의 모든 파일에 대해 반복
for filename in os.listdir(folder_path):
    if filename.endswith(".mp3"):  # MP3 파일인 경우
        mp3_path = os.path.join(folder_path, filename)
        new_filename = f"{singer_name}_{filename.replace('.mp3', '.wav')}"
        wav_path = os.path.join(output_path, new_filename)

        # MP3 파일을 불러와 WAV로 변환
        audio = AudioSegment.from_mp3(mp3_path)
        audio.export(wav_path, format="wav")
```

## audio slice

```python
import os
os.chdir('/audio-slicer') # 경로 설정

# !pip install -r requirements.txt

import sys
import glob
import librosa
import soundfile as sf

# slicer.py가 위치한 디렉토리 경로를 추가합니다.
sys.path.append('/content/drive/MyDrive/diff/preprocessing/4split/audio-slicer')

# slicer.py에서 Slicer 클래스를 임포트합니다.
from slicer2 import Slicer  # 파일 이름이 slicer2.py인지 확인하세요.

# 원본 파일 경로와 분할된 파일을 저장할 경로 설정
target_path = '/content/drive/Shareddrives/capstone/data/sikyung/sliced'

os.makedirs(target_path, exist_ok=True)

# 해당 폴더가 없으면 생성
if not os.path.exists(target_path):
    os.makedirs(target_path)

# 모든 wav 파일 목록을 가져옴
file_list = glob.glob(os.path.join(output_path, '*.wav'))

# Slicer 인스턴스 생성 (파라미터를 Slicer 클래스 생성자에 맞게 조정)
slicer = Slicer(sr=22050, threshold=-40, min_length=15000, min_interval=300, hop_size=20, max_sil_kept=500)

# 각 파일에 대해 처리
for file in file_list:
    # 오디오 로드
    audio, sr = librosa.load(file, sr=None)

    # 오디오 분할
    chunks = slicer.slice(audio)

    # 분할된 오디오 저장
    for i, chunk in enumerate(chunks):
        sf.write(os.path.join(target_path, f'{os.path.basename(file).rsplit(".", 1)[0]}_part{i}.wav'), chunk, sr)
```

## 파일 압축

```python
import os
import zipfile

# 압축할 파일들이 있는 폴더 경로
wav_folder_path = "/content/drive/Shareddrives/capstone/data/sikyung/sliced"

# 압축 파일을 저장할 경로와 파일명
zip_file_path = "/content/drive/Shareddrives/capstone/data/sikyung/sikyung.zip"

# ZIP 파일 생성
with zipfile.ZipFile(zip_file_path, 'w') as zipf:
    # 폴더 내의 모든 파일에 대해 반복
    for foldername, subfolders, filenames in os.walk(wav_folder_path):
        for filename in filenames:
            if filename.endswith(".wav"):  # WAV 파일인 경우
                # WAV 파일의 전체 경로
                file_path = os.path.join(foldername, filename)
                # ZIP 파일에 추가
                zipf.write(file_path, os.path.relpath(file_path, wav_folder_path))
```

```python
import os
import wave

directory = '/content/drive/Shareddrives/capstone/data/splitted'
files = os.listdir(directory)

def check_wav_file(file_path):
    try:
        with wave.open(file_path, 'rb') as wav_file:
            length = wav_file.getnframes()
            rate = wav_file.getframerate()
            channels = wav_file.getnchannels()
            sampwidth = wav_file.getsampwidth()
            print(f"File: {file_path}")
            print(f"Length: {length} frames")
            print(f"Channels: {channels}")
            print(f"Sample Width: {sampwidth} bytes")
            print(f"Frame Rate: {rate} frames/second")
            print("This appears to be a valid WAV file.\n")
    except wave.Error as e:
        print(f"File: {file_path}")
        print(f"Error: {e}")
        print("This file may not be a valid WAV file or may be corrupted.\n")

for file_name in files:
    file_path = os.path.join(directory, file_name)
    if os.path.exists(file_path):
        check_wav_file(file_path)
    else:
        print(f"File not found: {file_path}")
```
